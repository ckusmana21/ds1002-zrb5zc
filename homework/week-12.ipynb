## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

slices = [56, 66, 24, 124, 47]
labels = ["US", "UK", "MX", "CA", "JP"]

plt.pie(slices, labels=labels)
plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)

plt.hist(n, bins=25, edgecolor='black')
plt.xlabel('Numbers')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)

df = pd.DataFrame({'x': x, 'y': y})

sns.regplot(x='x', y='y', data=df)
plt.title('Scatterplot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
